```python
#PERCEPTRON LEARNING ALGORITHM 

import numpy as np

class Perceptron(object):
   def __init__(self, rate = 0.01, niter = 5):
      self.rate = rate
      self.niter = niter

   def fit(self, X, y):
      """Fit training data
      X : Training vectors, X.shape : [#samples, #features]
      y : Target values, y.shape : [#samples]
      """

      # weights
      self.weight = np.zeros(1 + X.shape[1])

      # Number of misclassifications
      self.errors = []  # Number of misclassifications

      for i in range(self.niter):
         err = 0
         for xi, target in zip(X, y):
            delta_w = self.rate * (target - self.predict(xi))
            self.weight[1:] += delta_w * xi
            self.weight[0] += delta_w
            err += int(delta_w != 0.0)
         self.errors.append(err)
      return self

   def net_input(self, X):
      """Calculate net input"""
      return np.dot(X, self.weight[1:]) + self.weight[0]

   def predict(self, X):
      """Return class label after unit step"""
      return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
from matplotlib.colors import ListedColormap
def plot_decision_regions(X, y, classifier, resolution=0.02):
   # setup marker generator and color map
   markers = ('s', 'x', 'o', '^', 'v')
   colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
   cmap = ListedColormap(colors[:len(np.unique(y))])

   # plot the decision surface
   x1_min, x1_max = X[:,  0].min() - 1, X[:, 0].max() + 1
   x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
   xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
   np.arange(x2_min, x2_max, resolution))
   Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
   Z = Z.reshape(xx1.shape)
   plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
   plt.xlim(xx1.min(), xx1.max())
   plt.ylim(xx2.min(), xx2.max())

   # plot class samples
   for idx, cl in enumerate(np.unique(y)):
      plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
      alpha=0.8, c=cmap(idx),
      marker=markers[idx], label=cl)
```

Let us create your own target function f and data set D and see how the perceptron learning algorithm works. Take d=2 so you can visualize the problem and choose a random line in the plane as your target function, where one side of the line maps to +1 and the other maps to -1. Choose the input xn of the data set as random points in the plane, and evaluate the target function on each xn to get the corresponding output yn. 

```python
#import the packages
import numpy as np
import pandas as pd
from functools import partial
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#function for generating random points
def generate_random_points(n, dim, num_grid_points, lb, ub):
  random_int = np.random.randint(num_grid_points, size=(n, dim))
  initial_lb = 0
  data = (random_int - initial_lb)/(num_grid_points-1-initial_lb)
  result = lb + (ub-lb)*data
  return result 
```

```python
def generate_binary_classes(n, dim, true_f, pred_f):
  class1, class2 = [], []
  while True:
    g = pred_f(1, dim).flatten()
    if true_f(g)>0 and len(class1) < n:
      class1.append(g)
    elif true_f(g) < 0 and len(class2) < n:
      class2.append(g)
    if(len(class1)==n and len(class2)==n):
      break
  
  return np.asarray(class1), np.asarray(class2)
```

```python
def true_function(x, coeffs):
  return coeffs.flatten()[0] + np.dot(coeffs.flatten()[1:], x.flatten())
```

```python
def generate_dataframe(n, dim, true_f, pred_f):
  class1, class2 = generate_binary_classes(n/2, dim, true_f, pred_f)
  cols = ['x' + str(i) for i in range(1, dim+1)] #columns x1, x2 ....

  df1 = pd.DataFrame(class1, columns=cols)
  df1['y'] = 1

  df2 = pd.DataFrame(class2, columns=cols)
  df2['y'] = -1

  df = pd.concat([df1, df2])
  df['x0'] = 1 #bias column
  df = df[['x0'] + cols + ['y']]
  return df
```

```python
def plot_graph(x1, df, w, g, lb, ub):
        figsize = (8,6)
        f, ax = plt.subplots(1, 1, figsize=figsize)
        cls1_df = df.loc[df['y']==1]
        cls2_df = df.loc[df['y']==-1]
        line = ax.plot(x1, -(w[0]+w[1]*x1), label='True Function')

        pluses = ax.scatter(cls1_df[['x1']].values, cls1_df[['x2']].values, marker='o', c= 'b', label='(+1) labels')
        minuses = ax.scatter(cls2_df[['x1']].values, cls2_df[['x2']].values, marker='o', c= 'g', label='(-1) labels')

        if g is not None:
            hypothesis = ax.plot(x1, -(g[0]+g[1]*x1), c = 'black', label='Hypothesis function')

        ax.set_ylabel("x2", fontsize=11)
        ax.set_xlabel("x1", fontsize=11)
        ax.set_title('Data set size = %s'%N, fontsize=9)
        ax.axis('tight')
        legend_x = 2.0
        legend_y = 0.5
        #ax.legend(loc='upper right')
        ax.legend(['True function', 'Hypothesis function', '(+1) labels', '(-1) labels',], loc='center right', bbox_to_anchor=(legend_x, legend_y))
        ax.set_ylim(bottom=lb, top=ub)
        plt.show()
```

```python
def PLA(points, dim, max_iter=1000, use_adaline=False, eta=1, isRandomized=False):
  w = np.zeros(dim+1) #initial value of weights
  xs, ys = points[:, :dim+1], points[:,dim+1]
  num_points = points.shape[0]

  #gradient descent
  for i in range(max_iter):
    correct_pred = set()
    indexs = np.arange(num_points)

    if isRandomized:
      indexs = np.random.choice(np.arange(num_points), num_points, replace=False)
    
    for idx in indexs:
      x, y = xs[idx], ys[idx]
      t1 = np.dot(w.T, x)
      prod = t1*y

      if prod<-100:
        t1 = -100

      threshold = 1 if use_adaline else 0
      t1 = t1 if use_adaline else 0
      if prod<=threshold:
        w = w + eta * (y-t1)*x
        break 
      else:
        correct_pred.add(idx)

    if(len(correct_pred) == num_points):
      break

  c=0
  for x,y in zip(xs,ys):
    prod = np.dot(w.T, x)*y
    if prod>0:
      c+=1

  w = w/w[-1]
  
  #printing the values
  print("Final correctness:", c)
  print("Total number of iteration:", i)
  print("Final weights:", w)

  return w, i
```

```python
def perceptron_experiment(N, dim, lb, ub, num_grid_points, coeff_lb, coeff_ub, 
                              eta, maxit, use_adaline=False, randomize=False,
                              show_plot=True, test_N = 10000):
    
    rns = generate_random_points(N, dim, num_grid_points, lb, ub)
    rn_func = partial(generate_random_points, num_grid_points = num_grid_points, lb = lb, ub = ub)
    coeffs = generate_random_points(1, dim+1, num_grid_points, coeff_lb, coeff_ub)

    norm_coeffs = coeffs.flatten()/coeffs.flatten()[-1]
    true_func = partial(true_function, coeffs = norm_coeffs)
    
    df = generate_dataframe(N, dim, true_func, rn_func)
    test_df = generate_dataframe(test_N, dim, true_func, rn_func)
    
    x1 = np.arange(lb, ub, 0.01)
    norm_g, num_its = PLA(df.values, dim, maxit, use_adaline, 
                                 eta, randomize)
    
    if show_plot:
        print('True coeffs: ', norm_coeffs)
        plot_graph(x1, df, norm_coeffs, norm_g, lb, ub)
```

```python
lb, ub = -100, 100
N, dim = 20, 2 
num_grid_points = 2000
coeff_lb, coeff_ub = -10, 10
eta = 1
maxit = 100
use_adaline, randomize = False, False
perceptron_experiment(N, dim, lb, ub, num_grid_points,
                             coeff_lb, coeff_ub, eta, maxit, use_adaline, randomize)
```
